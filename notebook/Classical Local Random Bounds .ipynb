```julia
using Plots; pyplot();
using LinearAlgebra;
```

```julia
SB1 = [1 0;0 1;0 0]
SA1 = [1 0 1;0 1 0]

SB2 = [0 0;0 1;1 0]
SA2 = [0 0 1;1 1 0]

S11 = SB1*SA1
S12 = SB1*SA2
S21 = SB2*SA1
S22 = SB2*SA2
```

```julia
n = 5000 # number of samples

# enumerated data series
μs = range(0, stop=1, length=n)
νs = range(0, stop=1, length=n)

# data arrangement for plotting
μ_grid = repeat(μs', n, 1)
ν_grid = repeat(νs, 1, n)

λ = zeros(n,n)


# computing weights from analytic functions
for i in 1:n
    μ = μs[i]

    for j in 1:n
        ν = νs[j]

        λ[j,i] = tr((ν*SB1 + (1-ν)*SB2)*(μ*SA1 + (1-μ)*SA2)) -2 
    end
    
end
```

```julia
colorbar_range = collect(-1:0.001:0)

plt = contourf(μ_grid, ν_grid, λ, title="Trajectories Across Local Random Slice",levels=20, xlabel="μ", ylabel="ν", zcolor=colorbar_range)
```
